```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Name": ['Ram', 'Shyam', 'Sagar', 'Hari', 'Gita', 'Kiran', 'Rita', 'Mohan', 'Sita', 'Anil', 'Nina', 'Arjun'],
    "Age": [22, 24, 20, 23, 21, 22, 24, 20, 22, 23, 21, 24],
    "Score": [85, 90, 78, 88, 92, 80, 87, 75, 89, 95, 84, 88],
    "Gender": ['Male', 'Male', 'Male', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male'],
    "City": ['Kathmandu', 'Pokhara', 'Lalitpur', 'Kathmandu', 'Bhaktapur', 'Pokhara', 'Kathmandu', 'Lalitpur', 'Bhaktapur', 'Kathmandu', 'Pokhara', 'Kathmandu'],
    "Passed": [True, True, False, True, True, True, True, False, True, True, True, True]
}


df = pd.DataFrame(data)
print(df)
```

```python
# Generate data for a heart shape
t = np.linspace(0, 2 * np.pi, 100)
x = 16 * np.sin(t) ** 3
y = 13 * np.cos(t) - 5 * np.cos(2 * t) - 2 * np.cos(3 * t) - np.cos(4 * t)

# Plot the scatter plot
plt.scatter(x, y, s=10, color='red')
plt.grid(True)
plt.axis('equal')  # Keep aspect ratio equal for better shape representation
plt.show()
```

```python
# Spiral pattern
theta = np.linspace(0, 4 * np.pi, 200)
r = theta  # Radial distance increases with angle
x = r * np.cos(theta)
y = r * np.sin(theta)

# Plot
plt.scatter(x, y, s=100, color='blue')
plt.grid(True)
plt.axis('equal')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig, axis = plt.subplots(2, 2, figsize=(6, 5))

# Spiral pattern (first subplot)
theta = np.linspace(0, 4 * np.pi, 200)
r = theta  # Radial distance increases with angle
x = r * np.cos(theta)
y = r * np.sin(theta)
axis[0, 0].scatter(x, y, s=100, color='blue')
axis[0, 0].set_title('Spiral')

# Circle pattern (second subplot)
theta = np.linspace(0, 2 * np.pi, 100)
x = 50 + 30 * np.cos(theta)  # Center at (50, 50) with radius 30
y = 50 + 30 * np.sin(theta)
axis[0, 1].scatter(x, y, s=100, color='red')
axis[0, 1].set_title('Circle')

# Heart pattern (third subplot)
t = np.linspace(0, 2 * np.pi, 100)
x = 16 * np.sin(t) ** 3
y = 13 * np.cos(t) - 5 * np.cos(2 * t) - 2 * np.cos(3 * t) - np.cos(4 * t)
axis[1, 0].scatter(x, y, s=100, color='green')
axis[1, 0].set_title('Heart')

# Lissajous Curve (fourth subplot)
t = np.linspace(0, 2 * np.pi, 500)
x = np.sin(3 * t)
y = np.sin(5 * t)
axis[1, 1].scatter(x, y, s=50, color='purple')
axis[1, 1].set_title('Lissajous')

# Adjust layout
plt.tight_layout()
plt.show()
```
