#  What percent of the total number of passengers on the Titanic belonged to each age group
 (e.g., child, adult, senior)?

```python



bins = [0, 18, 64, 100]  # Intervals: [0-18), [18-64), [64-100]
labels = ['child', 'adult', 'senior']  # Labels for the intervals

# Categorize ages
df['age_group'] = pd.cut(df['Age'], bins=bins, labels=labels)

df
```

```python
total_passengers = df.shape[0]
print(f"Total number of passengers {total_passengers}")
```

```python
age_group_count = df.groupby('age_group')['Name'].count()
age_group_count
```

```python

total_passengers = df.shape[0]


age_group_proportion = age_group_count / total_passengers

print(age_group_proportion)
```

```python
child_passenger = len(df[df['age_group'] == 'child'])
adult_passenger = len(df[df['age_group'] == 'adult'])
senior_passenger = len(df[df['age_group'] == 'senior'])

child_proportion = (child_passenger / total_passengers) * 100
adult_proportion = (adult_passenger / total_passengers) * 100
senior_proportion = (senior_passenger / total_passengers) * 100


print(f"Child Proportion: {child_proportion:.2f}%")
print(f"Adult Proportion: {adult_proportion:.2f}%")
print(f"Senior Proportion: {senior_proportion:.2f}%")
```
